```python
from datetime import date
d_init=pd.to_datetime(2/28/2017)
d_rec=media_data.Date
diff=d_rec - d_init
media_data['Day']=diff
```

```python
media_data['Day']=media_data['Day'].astype(str)
media_data['Day']=media_data['Day'].map(lambda x: x[0:5])
media_data['Day']=media_data['Day'].astype(int)
```

```python
media_data.head()
```

```python
media_data.columns
```

```python
media_data.columns
```

```python
media_data['Day']=media_data['Day']-17225
```

```python
media_data['Day_number']=(media_data['Day']+3)%7
media_data.Day_number.replace(0,7,inplace=True)
media_data['Day_number']=media_data['Day_number'].astype(int)
```

```python
media_data
```

```python
def func(i):
    if i%7 == 5: return 1
    elif i%7 ==4: return 1
    else: return 0
    
media_data['Weekend']=[func(i) for i in media_data['Day']]
```

```python
#Days-Views + #Days-Ad_impressions

fig=plt.figure()
host=fig.add_subplot(111)

par1=host.twinx()
par2=host.twinx()

color1=plt.cm.viridis(0)
color2=plt.cm.viridis(0.5)
color3=plt.cm.viridis(.9)

p1,=host.plot(media_data.Day,media_data.Views_show, color=color1,label='Views')
p2,=par1.plot(media_data.Day,media_data.Ad_impression, color=color2, label='Ad Impression')

lns=[p1,p2]
host.legend(handles=lns, loc='best')

par2.spines['right'].set_position(('outward',60))
par2.xaxis.set_ticks([])

host.yaxis.label.set_color(p1.get_color())
par1.yaxis.label.set_color(p2.get_color())

plt.savefig("pyplot_multiple.png",bbox_inches='tight')
```
